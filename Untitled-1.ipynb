```python
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
import re

def create_distribution(response, frequency, suptitle):
    total = sum(frequency)
    percentages = [(freq / total * 100) for freq in frequency]

    cmap = get_cmap('tab20c')
    colors = cmap(range(len(response)))

    _, (ax1, ax2) = plt.subplots(2,1, figsize=(8, 12))
    ax1.pie(percentages, labels=response, autopct='%1.1f%%', startangle=90, colors=colors, counterclock=True, wedgeprops={'edgecolor': 'gray'})
    ax1.set_title('Pie Chart')
    ax1.legend(response, loc='upper left', bbox_to_anchor=(0.95, 1), title="Response", fontsize=8)
    
    
    ax2.bar(response, frequency, color='skyblue') 
    ax2.set_ylabel('Frequency (n)')
    ax2.set_xlabel('Response')
    ax2.set_title('Bar Chart')

    for i, freq in enumerate(frequency):
        ax2.text(response[i], freq + 1, str(freq), ha='center', va='bottom')

    plt.xticks(rotation=0)

    sanitized_filename = re.sub(r'[\/:*?<>|]', '_', suptitle)
    file = f"{sanitized_filename}"
    plt.suptitle(suptitle, fontsize=16)
    plt.tight_layout()
    plt.savefig(f"images/{sanitized_filename}")
    plt.show()

```
